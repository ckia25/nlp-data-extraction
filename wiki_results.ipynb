```python
import pandas as pd
import numpy as np
import os
import sys
dir_name = 'wiki_generations'
dfs = []
for file in os.listdir(dir_name):
    path = f"{dir_name}/{file}"
    dfs.append(pd.read_csv(path, index_col=0, encoding_errors='ignore')[:500])
df = pd.concat(dfs)
print(df.shape)
df.to_csv('wiki_combined_results.csv')

```

# $\geq$ Thresholds

```python
import pandas as pd
def count_geq_1(series):
    return (series >= 1).sum()
def count_geq_2(series):
    return (series >= 2).sum()
def count_geq_3(series):
    return (series >= 3).sum()
def count_geq_4(series):
    return (series >= 4).sum()
def count_geq_5(series):
    return (series >= 5).sum()
def count_geq_6(series):
    return (series >= 6).sum()
def count_geq_7(series):
    return (series >= 7).sum()
def order(series):
    if series == '160m':
        return 0
    if series == '410m':
        return 1
    if series == '1b':
        return 2
    if series == '1.4b':
        return 3
    if series == '2.8b':
        return 4
    return None

df = pd.read_csv('wiki_combined_results.csv')
df2 = df.drop(df.columns[0], axis=1)
df3 = df2.sort_values('Example Id')[:-4]
df4 = df3.groupby(['model', 'Prefix Length']).agg(
    Mean=('Matching Token Count', 'mean'), 
    Max=('Matching Token Count', 'max'),
    Sum=('Matching Token Count', 'sum'),
    Freq_GEQ_1=('Matching Token Count', count_geq_1),
    Freq_GEQ_2=('Matching Token Count', count_geq_2),
    Freq_GEQ_3=('Matching Token Count', count_geq_3),
    Freq_GEQ_4=('Matching Token Count', count_geq_4),
    Freq_GEQ_5=('Matching Token Count', count_geq_5),
    Freq_GEQ_6=('Matching Token Count', count_geq_6),
    Freq_GEQ_7=('Matching Token Count', count_geq_7), 
    model_id=('model', 'first')
)
df4['model_id'] = df4['model_id'].apply(order)
display(df4.sort_values(['Prefix Length', 'model_id']))
display(df4.sort_values(['model_id']))
```
